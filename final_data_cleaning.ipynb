I found this code from here: https://stackoverflow.com/questions/75618004/concatenate-all-csv-files-in-a-folder-using-pandas-and-remove-leading-comma

```python
import glob
from tkinter.filedialog import askdirectory
import os
```

```python
#create a list of all directory paths:
path_list = ['1_January',
             '2_February',
             '3_March',
             '4_April',
             '5_May',
             '6_June',
             '7_July',
             '8_August',
             '9_September',
             '10_October',
             '11_November',
             '12_December'
]
```

```python
print(os.getcwd())
```

```python
os.chdir('2022-citibike-tripdata')
```

```python
print(os.getcwd())
```

```python
all_files = []
for path in path_list:
    #navigate to child directory
    os.chdir(path)
    #grab all csv file names
    files = glob.glob("*.csv")
    #construct the file names 
    all_files.append([path + '/' + f for f in files])
    #now we return to parent directory so we can switch child directories
    os.chdir(os.path.abspath('..'))

#flattens the list
all_files = [item for sublist in all_files for item in sublist]
print(all_files)
```

```python
#cwd should be 2022-citibike-tripdata

print(os.getcwd())
```

```python
flist = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=False)
    flist.append(df)

df_out = pd.concat(flist, axis=0, ignore_index=False)
display(df_out.head())
```
